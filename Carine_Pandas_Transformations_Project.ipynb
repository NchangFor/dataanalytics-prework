## 3. Is there a difference of average room space (`Space` divided by `Rooms`) for apartments built after 2000 versus apartment built before 2000?

```
Immo["Average_Room_space"] = (Immo.Space / Immo.Rooms).round(1)
Immo.head()
```

```
Immo_After_2000 = Immo[Immo.Year_Construction > 2000]
Immo_After_2000.head()
```

```
Immo_Before_2000 = Immo[Immo.Year_Construction < 2000]
Immo_Before_2000.head()
```

```
Sample_A = Immo_After_2000.Average_Room_space
Sample_A.head()
```

```
Sample_B = Immo_Before_2000.Average_Room_space
Sample_B.head()
```

```
print("Null_Hypothesis = (Sample_A = Sample_B)")
```

```
t_results = stats.ttest_ind(Sample_A, Sample_B)
t_results
```

```
alpha = 0.1
print(alpha)
```

```
if (t_results[1] < alpha):
  print("Sample_A and Sample_B are different")
```

## 4. How many apartments were posted each month? 



```
Immo["posted_date"] = np.random.choice(pd.date_range('2021-01-01', '2021-12-31'), len(Immo))
Immo["Month"] = Immo.posted_date.dt.month
```

```
Immo.head()
```

```
Monthly_Apartment_Posted = Immo.groupby('Month')[["Rooms"]].count()
Monthly_Apartment_Posted.head()
```

Second Option

```
Yearly_Apartment_Posted = Immo.groupby('Year_Construction')[["Rooms"]].count()
Yearly_Apartment_Posted.tail()
```

```
Monthly_Apartment_Posted = (Yearly_Apartment_Posted/12).round(1)
Monthly_Apartment_Posted.tail()
```
